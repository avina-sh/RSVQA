## Objective of the Model
<br>
The primary objective of this model is to provide answers to open-ended questions about images, particularly for remote sensing data. It achieves this by extracting meaningful features from the image and question, fusing them, and classifying the answer from a set of possible choices. This model is designed to make remote sensing data more accessible and actionable by enabling users to ask questions about the data and receive informative answers.
<br>
<br>

## Layered Architecture

The architecture of the VQA model described in the provided text can be divided into several layers, with each layer serving a specific purpose in the overall model. Here's the layered architecture of the model:

### Feature Extraction:

* **Visual Part:**
    * **Input:** 2-D image
    * **Feature Extraction:** A ResNet-50 model pre-trained on ImageNet is used as the Convolutional Neural Network (CNN) to extract visual features from the image.
    * **Last Layers:** Two fully-connected layers reducing ResNet output size of 8192 to 2048 and from 2048 to 1200
    * **Output:** A 1200-D vector representing the visual features of the image.

* **Language Part:**
    * **Input:** A natural language question
    * **Feature Extraction:** The skip-thoughts model, a recurrent neural network trained on the BookCorpus dataset, encodes the question into a vector representation.
    * **Output:** A feature vector obtained from the skip-thoughts model, followed by a fully connected layer, resulting in a 1200-D vector representing the question.

### Fusion:

   * **Input:** The 1200-D feature vectors from both the visual and language parts.
   * **Fusion Operation:** A pointwise multiplication is applied element-wise to the vectors after applying the hyperbolic tangent (tanh) function to the elements.
   * **Output:** A single 1200-D vector that serves as a joint embedding of both the image and the question. This vector is used as input for the prediction step.


### Prediction:

   * **Input:** The 1200-D fused vector obtained from the fusion step.
   * **Prediction Model:** A Multi-Layer Perceptron (MLP) with one hidden layer consisting of 256 elements.
   * **Task:** The problem is formulated as a classification task, where each possible answer is treated as a class.
   * **Output:** The size of the output vector depends on the number of possible answers, with each element in the vector representing the likelihood of a particular answer.


## Definition of the data loader and encoder
<br>

```python
# Import libraries for experiment tracking, JSON data processing, and progress bar visualization
from comet_ml import Experiment  # Comet.ml for experiment tracking and visualization
import json  # Module for working with JSON data
from tqdm import tqdm  # Library for creating progress bars during iterations


MAX_ANSWERS = 1000
LEN_QUESTION = 20

class VocabEncoder():
    
    """
    VocabEncoder is a class designed to process and encode textual data, specifically questions or answers,
    creating a vocabulary based on word frequencies. It offers methods for encoding and decoding sentences
    using this vocabulary.

    Attributes:
    - encoder_type (str): Indicates whether the encoder is for 'questions' or 'answers'.
    - questions (bool): Flag to determine whether the encoder is for questions (True) or answers (False).
    - words (dict): A dictionary containing words from the dataset and their corresponding indices based on frequency.
                   Includes a special token '<eos>' for end-of-sentence.
    - list_words (list): A list representation of the vocabulary, including the special token '<eos>'.

    Methods:
    - __init__(JSONFile, questions=True): Initializes the VocabEncoder with data from a JSON file and the type of sentences.
    - encode(sentence): Encodes a given sentence into a sequence of indices based on the vocabulary.
    - getVocab(): Returns the list of words in the vocabulary.
    - decode(sentence): Decodes a sequence of indices back into a sentence using the vocabulary.
    """

    def __init__(self, JSONFile, questions=True):
        # Initialize the VocabEncoder with the specified JSON file and whether it is for questions or answers
        self.encoder_type = 'answer'
        if questions:
            self.encoder_type = 'question'
        self.questions = questions

        # Initialize a dictionary to store word frequencies
        words = {}

        # Read the JSON file and extract the relevant type of sentences (questions/answers)
        print(f"Reading {JSONFile}..", flush=True, end="")
        with open(JSONFile) as json_data:
            data = json.load(json_data)[self.encoder_type + 's']
        print("Done.")

        # Parse the data to count word frequencies
        print(f"Parsing {JSONFile}..", flush=True, end="")
        for i in tqdm(range(len(data))):
            if data[i]["active"]:
                sentence = data[i][self.encoder_type]
                # Remove punctuation at the end of the sentence
                if sentence[-1] == "?" or sentence[-1] == ".":
                    sentence = sentence[:-1]

                # Tokenize the sentence and update word frequencies
                tokens = sentence.split()
                for tok in tokens:
                    tok = tok.lower()
                    if tok not in words:
                        words[tok] = 1
                    else:
                        words[tok] += 1

        # Free the memory by releasing the data
        data = None
        print("Done.")

        # Sort words by frequency and create a vocabulary
        sorted_words = sorted(words.items(), key=lambda kv: kv[1], reverse=True)
        print(f"List of words found: {sorted_words}")
        self.words = {'<eos>': 0}
        self.list_words = ['<eos>']
        for i, word in enumerate(sorted_words):
            if self.encoder_type == 'answer':
                if i >= MAX_ANSWERS:
                    break
            self.words[word[0]] = i + 1
            self.list_words.append(word[0])

            
    def encode(self, sentence):
        
        """
        Encodes a given sentence into a sequence of indices based on the vocabulary.

        Parameters:
        - sentence (str): The input sentence to be encoded.

        Returns:
        - res (list): A list of indices representing the encoded sentence.
        """
        
        res = []
        if sentence[-1] == "?" or sentence[-1] == ".":
            sentence = sentence[:-1]

        tokens = sentence.split()
        for tok in tokens:
            tok = tok.lower()
            res.append(self.words[tok] + 1)

        res.append(self.words['<eos>'] + 1)

        # Pad the sequence with zeros if it's a question
        if self.questions:
            while len(res) < LEN_QUESTION:
                res.append(0)
            res = res[:LEN_QUESTION]
        return res

    
    def getVocab(self):
        
        """
        Returns the list of words in the vocabulary.

        Returns:
        - list_words (list): A list of words in the vocabulary.
        """
        
        return self.list_words

    
    def decode(self, sentence):
        
        """
        Decodes a sequence of indices back into a sentence using the vocabulary.

        Parameters:
        - sentence (list): The input sequence of indices to be decoded.

        Returns:
        - res (str): The decoded sentence.
        """
        
        res = ""
        for i in sentence:
            if self.list_words[i - 1] == "<eos>":
                break
            res += self.list_words[i - 1]
            res += " "
        res = res[:-1]  # Remove the last space
        if self.questions:
            res += "?"
        return res
```

```python
# Import necessary libraries for file operations, data processing, and visualization
import os.path  # Module for working with file paths
import json  # Module for working with JSON data
import random  # Module for random number generation
from tqdm import tqdm  # Library for creating progress bars
import numpy as np  # NumPy library for numerical operations
from skimage import io  # Module for reading and writing images
from torch.utils.data import Dataset  # PyTorch's Dataset class for creating custom datasets
import torchvision.transforms as T  # Transformations for image data


class VQALoader(Dataset):
    """
    VQALoader is a PyTorch Dataset class for loading and processing Visual Question Answering (VQA) data.

    Attributes:
    - imgFolder (str): Path to the folder containing images.
    - images_file (str): Path to the JSON file containing image information.
    - questions_file (str): Path to the JSON file containing question information.
    - answers_file (str): Path to the JSON file containing answer information.
    - encoder_questions (VocabEncoder): Encoder for questions.
    - encoder_answers (VocabEncoder): Encoder for answers.
    - train (bool): Flag indicating whether the dataset is for training.
    - ratio_images_to_use (float): Ratio of images to use from the total available.
    - transform (torchvision.transforms): Image transformation.
    - number_outputs (int): Number of most frequent answers to consider during training.
    - selected_answers (list): List of selected answers during testing.
    - patch_size (int): Size of image patches.

    Methods:
    - __init__(imgFolder, images_file, questions_file, answers_file, encoder_questions, encoder_answers, train=True, ratio_images_to_use=1, transform=None, number_outputs=1000, selected_answers=None, patch_size=120):
        Initializes the VQALoader with the specified parameters.

    - __len__():
        Returns the number of samples in the dataset.

    - __getitem__(idx):
        Returns a sample from the dataset based on the provided index.
    """
    def __init__(self, imgFolder, images_file, questions_file, answers_file, encoder_questions, encoder_answers, train=True, ratio_images_to_use=1, transform=None, number_outputs=1000, selected_answers=None, patch_size=120):
        # Initialize the dataset attributes
        self.transform = transform
        self.encoder_questions = encoder_questions
        self.encoder_answers = encoder_answers
        self.train = train
        self.imgFolder = imgFolder

        # Load JSON files containing image, question, and answer information
        print("Loading JSONs...", flush=True, end="")
        with open(questions_file) as json_data:
            questionsJSON = json.load(json_data)

        with open(answers_file) as json_data:
            answersJSON = json.load(json_data)

        with open(images_file) as json_data:
            imagesJSON = json.load(json_data)
        print("Done.", flush=True)

        # Extract active image IDs based on the ratio to use
        images = [img['id'] for img in imagesJSON['images'] if img['active']]
        images = images[:int(len(images) * ratio_images_to_use)]
        self.img_ids = images

        # Construct the dataset based on training or testing conditions
        print("Construction of the Dataset", flush=True, end="")
        if train:
            # Initialize a frequency dictionary for answer occurrences
            self.freq_dict = {}
            for i, image in enumerate(tqdm(images)):
                for questionid in imagesJSON['images'][image]['questions_ids']:
                    question = questionsJSON['questions'][questionid]

                    # Get the answer string for the first answer ID
                    answer_str = answersJSON['answers'][question["answers_ids"][0]]['answer']

                    # Update the frequency dictionary
                    if answer_str not in self.freq_dict:
                        self.freq_dict[answer_str] = 1
                    else:
                        self.freq_dict[answer_str] += 1

            # Sort the frequency dictionary by occurrences in descending order
            self.freq_dict = sorted(self.freq_dict.items(), key=lambda x: x[1], reverse=True)
            self.selected_answers = []
            self.non_selected_answers = []
            coverage = 0
            total_answers = 0

            # Select the top 'number_outputs' most frequent answers
            for i, key in enumerate(self.freq_dict):
                if i < number_outputs:
                    self.selected_answers.append(key[0])
                    coverage += key[1]
                else:
                    self.non_selected_answers.append(key[0])
                total_answers += key[1]

            print(f"The {number_outputs} most frequent answers cover {coverage/total_answers*100}% of the total answers.")
        else:
            self.selected_answers = selected_answers

        # Initialize image arrays and question-answer pairs
        self.images = np.empty((len(images), patch_size, patch_size, 3), dtype=np.uint8)
        self.images_questions_answers = []
        for i, image in enumerate(tqdm(images)):
            # Load image data into the array
            self.images[i, :, :, :] = io.imread(os.path.join(self.imgFolder, str(image) + '.tif'))

            # Extract information for each question associated with the image
            for questionid in imagesJSON['images'][image]['questions_ids']:
                question = questionsJSON['questions'][questionid]

                question_str = question["question"]
                type_str = question["type"]
                answer_str = answersJSON['answers'][question["answers_ids"][0]]['answer']

                # Create question-answer pairs based on selected answers (for testing)
                if self.selected_answers is None:
                    self.images_questions_answers.append([self.encoder_questions.encode(question_str),
                                                          self.encoder_answers.encode(answer_str),
                                                          i, image, type_str])
                else:
                    if answer_str in self.selected_answers:
                        self.images_questions_answers.append([self.encoder_questions.encode(question_str),
                                                              self.selected_answers.index(answer_str),
                                                              i, image, type_str])
        print("Done.")

    def __len__(self):
        """
        Returns the number of samples in the dataset.

        Returns:
        - len (int): The number of samples in the dataset.
        """
        return len(self.images_questions_answers)

    def __getitem__(self, idx):
        """
        Returns a sample from the dataset based on the provided index.

        Parameters:
        - idx (int): Index of the sample.

        Returns:
        - Sample (tuple): Tuple containing question encoding, answer encoding, and image data.
        """
        question = self.images_questions_answers[idx]
        img = self.images[question[2]]

        # Data augmentation for training images
        if self.train:
            if random.random() < .5:
                img = np.flip(img, axis=0)
            if random.random() < .5:
                img = np.flip(img, axis=1)
            if random.random() < .5:
                img = np.rot90(img, k=1)
            if random.random() < .5:
                img = np.rot90(img, k=3)

        # Apply image transformation if specified
        if self.transform:
            imgT = self.transform(img.copy())

        # Return different components of the sample based on training or testing
        if self.train:
            return np.array(question[0], dtype='int16'), np.array(question[1], dtype='int16'), imgT
        else:
            return np.array(question[0], dtype='int16'), np.array(question[1], dtype='int16'), imgT, question[4], T.ToTensor()(img / 255), question[3]
```

## Model definition

```python
# Importing necessary libraries
from torchvision import models as torchmodels  # torchvision library for pre-trained models
import torch.nn as nn  # PyTorch neural network module
import models.seq2vec  # Custom module for sequence-to-vector models
import torch.nn.functional as F  # PyTorch functional interface
import torch  # PyTorch library for tensor operations

# Constants defining the architecture
VISUAL_OUT = 2048
QUESTION_OUT = 2400
FUSION_IN = 1200
FUSION_HIDDEN = 256
DROPOUT_V = 0.5
DROPOUT_Q = 0.5
DROPOUT_F = 0.5

# Definition of the VQAModel class
class VQAModel(nn.Module):
    def __init__(self, vocab_questions, input_size=120, number_outputs=100):
        super(VQAModel, self).__init__()

        # Dropout layers for different components
        self.dropoutV = torch.nn.Dropout(DROPOUT_V)
        self.dropoutQ = torch.nn.Dropout(DROPOUT_Q)
        self.dropoutF = torch.nn.Dropout(DROPOUT_F)

        # Sequence-to-vector model for questions
        self.seq2vec = models.seq2vec.factory(vocab_questions, {'arch': 'skipthoughts', 'dir_st': 'data/skip-thoughts', 'type': 'BayesianUniSkip', 'dropout': 0.25, 'fixed_emb': False})
        for param in self.seq2vec.parameters():
            param.requires_grad = False

        # Linear layer for question embedding
        self.linear_q = nn.Linear(QUESTION_OUT, FUSION_IN)

        # Visual model using ResNet152
        self.visual = torchmodels.resnet152(pretrained=True)
        extracted_layers = list(self.visual.children())[:8]  # Remove the last fc and avg pool
        self.visual = torch.nn.Sequential(*extracted_layers)
        for param in self.visual.parameters():
            param.requires_grad = False

        # Linear layer for visual embedding
        self.visual_body = self.visual
        output_size = round((input_size / 32))**2
        self.visual_end = torch.nn.Conv2d(VISUAL_OUT, int(VISUAL_OUT/output_size), 1)
        self.linear_v = nn.Linear(VISUAL_OUT, FUSION_IN)

        # Classification layers
        self.linear_classif1 = nn.Linear(FUSION_IN, FUSION_HIDDEN)
        self.linear_classif2 = nn.Linear(FUSION_HIDDEN, number_outputs)

    def forward(self, input_v, input_q):
        # Visual input processing
        x_v = self.visual_body(input_v)
        x_v = self.visual_end(x_v)
        x_v = x_v.view(-1, VISUAL_OUT)
        x_v = self.dropoutV(x_v)
        x_v = self.linear_v(x_v)
        x_v = x_v.tanh()

        # Question input processing
        x_q = self.seq2vec(input_q)
        x_q = self.dropoutV(x_q)
        x_q = self.linear_q(x_q)
        x_q = x_q.tanh()

        # Fusion of visual and question features
        x = torch.mul(x_v, x_q)
        x = x.tanh()
        x = self.dropoutF(x)

        # Classification layers
        x = self.linear_classif1(x)
        x = x.tanh()
        x = self.dropoutF(x)
        x = self.linear_classif2(x)

        return x
```

```python
# Instantiating the VQAModel with the vocabulary from encoder_questions and specifying the number of output classes
model = VQAModel(encoder_questions.getVocab(), number_outputs=number_outputs)
```

## Training the model

```python
import matplotlib.pyplot as plt

hyper_params = {
    "optimizer":"Adam",
    "batch_size": 1024,
    "num_epochs": 10,
    "learning_rate": 10**-6,
    #For logging purposes
    "ratio_images_to_use": ratio_images_to_use,
    "number_outputs": number_outputs
}

if hyper_params['optimizer'] == "Adam":
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad,model.parameters()), lr=hyper_params['learning_rate'])
else:
    optimizer = torch.optim.SGD(filter(lambda p: p.requires_grad,model.parameters()), lr=hyper_params['learning_rate'])
criterion = torch.nn.CrossEntropyLoss(reduction='sum')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

model = model.to(device)

loaders = {}
for step in ["train", "validation"]:
    train = step == "train"
    loaders[step] = torch.utils.data.DataLoader(dataset[step], batch_size=hyper_params['batch_size'], shuffle=train, num_workers=0)          
```

```python
# Creating an Experiment for logging with Comet.ml
experiment = Experiment(project_name="vqaxben", workspace="sylvainlobry", log_code=True)

# Logging hyperparameters
experiment.log_parameters(hyper_params)

# Dictionary to store losses and number of iterations
losses = {"train": [], "validation": []}
number_iters = {"train": 0, "validation": 0}

# Training loop for epochs
for epoch in range(hyper_params['num_epochs']):
    print(f"Starting epoch {epoch}:")

    # Iterating over training and validation steps
    for step in ["train", "validation"]:
        # Setting the model to training mode for the train step, and evaluation mode for the validation step
        if step == "train":
            model.train()
        else:
            model.eval()

        total_correct = 0
        with torch.set_grad_enabled(step == "train"):
            losses[step].append(0)

            # Iterating over batches in the DataLoader
            for i, data in enumerate(tqdm(loaders[step])):
                if step == "train":
                    question, answer, image = data
                else:
                    question, answer, image, type_str, image_original, _ = data

                # Converting data to PyTorch tensors and moving to the specified device (GPU)
                question = torch.tensor(question).long().to(device)
                answer = torch.tensor(answer).long().to(device)
                image = torch.tensor(image).float().to(device)

                # Run the model and compute the loss
                pred = model(image, question)
                loss = criterion(pred, answer)

                # Backpropagation and optimization for the training step
                if step == "train":
                    optimizer.zero_grad()
                    loss.backward()
                    optimizer.step()

                # Accumulating the loss
                losses[step][epoch] += loss.cpu().item()

                # Compute the accuracy
                _, predicted = torch.max(pred.data, 1)
                batch_correct = (predicted == answer.data).sum()
                total_correct += batch_correct

                # Logging metrics for each batch
                experiment.log_metric(step+"_loss_batch", loss.cpu().item()/question.size(0), step=number_iters[step])
                experiment.log_metric(step+"_accuracy_batch", batch_correct/question.size(0), step=number_iters[step])
                number_iters[step] += 1

            # Computing average loss for the epoch
            losses[step][epoch] /= len(dataset[step])
            print(f"{step} loss = {losses[step][epoch]}")

            # Logging metrics for the entire epoch
            experiment.log_metric(step+"_loss", losses[step][epoch], epoch=epoch)
            experiment.log_metric(step+"_accuracy", total_correct/len(dataset[step]), epoch=epoch)

        # Plotting losses after the first epoch
        if epoch > 0:
            plt.figure()
            handles = []
            handles.append(plt.plot(losses["train"], label="train")[0])
            handles.append(plt.plot(losses["validation"], label="validation")[0])
            plt.legend(handles=handles)
            plt.show()

            # Saving the model checkpoint
            torch.save({"model_state_dict": model.state_dict(), "optimizer": optimizer.state_dict(), "losses": losses, "number_iters": number_iters, "epoch": epoch}, "Model"+experiment.get_key()+".tar")

# Ending the Comet.ml experiment
experiment.end()
```

## Evaluate Model Performance

```python
from sklearn.metrics import accuracy_score  # Importing accuracy_score from scikit-learn

# Calculating global accuracy
global_acc = accuracy_score(results["GTs"], results["preds"])

# Lists to store binary (Yes/No) predictions and ground truths, as well as multi-class (LC) predictions and ground truths
yn_preds = []
yn_gts = []
LC_preds = []
LC_gts = []

# Separating predictions and ground truths based on answer type (Yes/No or LC)
for i, ans in enumerate(results["GTs"]):
    if selected_answers[ans] == "yes" or selected_answers[ans] == "no":
        yn_preds.append(results["preds"][i])
        yn_gts.append(ans)
    else:
        LC_preds.append(results["preds"][i])
        LC_gts.append(results["GTs"][i])

# Calculating accuracy for Yes/No and LC separately
yn_acc = accuracy_score(yn_gts, yn_preds)
LC_acc = accuracy_score(LC_gts, LC_preds)

# Displaying accuracy results
print(f"Global accuracy: {'{:.2f}'.format(global_acc*100)}%")
print(f"Yes/No accuracy: {'{:.2f}'.format(yn_acc*100)}%")
print(f"LC accuracy: {'{:.2f}'.format(LC_acc*100)}%")
```

```python
from sklearn.metrics import confusion_matrix  # Importing confusion_matrix from scikit-learn
import matplotlib.pyplot as plt  # Importing pyplot from Matplotlib
import matplotlib.ticker as ticker  # Importing ticker for configuring axis ticks in Matplotlib
import matplotlib as mpl  # Importing Matplotlib as mpl for configuration purposes
import numpy as np  # Importing NumPy for numerical operations and array handling

# Computing the confusion matrix
mat = confusion_matrix(results["GTs"], results["preds"])

# Plotting the confusion matrix
number_to_plot = 29  # Number of classes to plot (adjust as needed)

# Creating a figure
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
cax = ax.matshow(np.log(mat + 1)[:number_to_plot, :number_to_plot], cmap="YlGn")

# Configuring colorbar
colorbar_ticks = np.array([0, 2, 4, 6, 8, 10, 12, 14])
cbar = fig.colorbar(cax)
cbar.set_ticks(colorbar_ticks)
cbar.set_ticklabels(['{:.0f}'.format(x) for x in np.exp(colorbar_ticks) - 1])

# Adjusting tick labels based on the number of classes
if number_to_plot < 30:
    ax.set_xticklabels([''] + selected_answers[:number_to_plot], rotation=45, ha="left")
    ax.set_yticklabels([''] + selected_answers[:number_to_plot])
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
else:
    ax.set_xticks([], [])
    ax.set_yticks([], [])

# Adding labels and displaying the plot
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Saving the figure as an SVG file
fig.savefig('confusion_matrix_' + str(number_to_plot) + '.svg')
```

```python
from PIL import Image, ImageDraw, ImageFont  # Importing necessary modules from the Python Imaging Library (PIL)
import matplotlib.pyplot as plt  # Importing pyplot from Matplotlib
%matplotlib inline

# Index of the result to visualize
index = 1188548

# Generating the path to the image file
image = os.path.join('Images', str(results['img_ids'][index]) + '.tif')

# Printing the image ID
print(results['img_ids'][index])

# Capitalizing and extracting question, ground truth answer, and predicted answer
question = results['questions'][index].capitalize()
answer = selected_answers[results['GTs'][index]].capitalize()
pred = selected_answers[results['preds'][index]].capitalize()

# Loading the template image
im = Image.open("TemplateAnswerVQA.png")
W, H = im.size

# Opening and resizing the satellite image
im_sat = Image.open(image)
im_sat = im_sat.resize((512, 512))

# Pasting the satellite image onto the template
im.paste(im_sat, (21, 21))

# Defining the box for the question and formatting it
box_question = [(21, 550), (531, 639)]
max_char = 50
question_formated = ""
for i in np.arange(0, len(question), max_char):
    if i > 0:
        question_formated += "\n"
    start = i
    end = i + 50
    if end > len(question):
        end = len(question)
    question_formated += question[start:end]

# Drawing the formatted question on the image
draw = ImageDraw.Draw(im)
font = ImageFont.truetype("micross.ttf", 20)
w, h = draw.multiline_textsize(question_formated, font=font, spacing=0)
to_put_w = box_question[0][0] + ((box_question[1][0] - box_question[0][0]) - w) / 2
to_put_h = box_question[0][1] + ((box_question[1][1] - box_question[0][1]) - h) / 2
draw.multiline_text((to_put_w, to_put_h), question_formated, fill="black", font=font, align="center")

# Adjusting font size for answer and predicted answer
fontsize = 30
if len(answer) >= 15 or len(pred) >= 15:
    fontsize = 20
font = ImageFont.truetype(os.path.join("/content/drive/My Drive/RSVQA/", "micross.ttf"), fontsize)

# Defining the box for the ground truth answer and drawing it on the image
box_gt = [(21, 640), (276, 727)]
w, h = draw.textsize(answer, font=font)
to_put_w = box_gt[0][0] + ((box_gt[1][0] - box_gt[0][0]) - w) / 2
to_put_h = box_gt[0][1] + ((box_gt[1][1] - box_gt[0][1]) - h) / 2
draw.multiline_text((to_put_w, to_put_h), answer, fill="black", font=font, align="center")

# Determining the fill color for the predicted answer based on correctness
if pred == answer:
    fill = "green"
else:
    fill = "red"

# Defining the box for the predicted answer and drawing it on the image
box_pred = [(277, 640), (531, 727)]
w, h = draw.textsize(pred, font=font)
to_put_w = box_pred[0][0] + ((box_pred[1][0] - box_pred[0][0]) - w) / 2
to_put_h = box_pred[0][1] + ((box_pred[1][1] - box_pred[0][1]) - h) / 2
draw.multiline_text((to_put_w, to_put_h), pred, fill=fill, font=font, align="center")

# Displaying the image using Matplotlib
fig = plt.figure(dpi=512)
plt.imshow(im)
plt.axis('off')
plt.show()

# Saving the figure as a PNG file in the Output directory
fig.savefig(os.path.join("Output", f"{index}.png"), bbox_inches=0)
```
